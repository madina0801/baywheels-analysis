```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
path = '/kaggle/input/baywheels-dataset'
files = [file for file in os.listdir(path)]

dfs = []
for file in files:
    df = pd.read_csv(os.path.join(path, file))
    dfs.append(df)

tripdata = pd.concat(dfs, ignore_index=True)
```

```python
tripdata["started_at"] = pd.to_datetime(tripdata["started_at"], format="mixed")
```

```python
tripdata['start_hour'] = tripdata["started_at"].dt.hour
tripdata['start_day'] = tripdata["started_at"].dt.day_name()
```

```python
tripdata["route"] = tripdata["start_station_name"] + "->" + tripdata["end_station_name"]
tripdata["route"].value_counts()
```

```python
tripdata["ride_duration_in_minutes"] = (pd.to_datetime(tripdata["ended_at"], format="mixed") - pd.to_datetime(tripdata["started_at"])).dt.total_seconds() / 60
tripdata["ride_duration_in_minutes"].apply(lambda x: round(x, 1))
```

```python
# Check for incorrect time stamps
negative_durations = tripdata[tripdata['ended_at'] < tripdata['started_at']]
# print(negative_durations) -> 74 records

# Delete the records with the incorrect time stamps
tripdata = tripdata[tripdata['ended_at'] >= tripdata['started_at']]
tripdata.shape
```

```python
tripdata.rename(columns={"rideable_type": "bike_type", "member_casual": "member_type"})
```

```python
tripdata.to_csv("baywheels_data.csv", index=False)
```
